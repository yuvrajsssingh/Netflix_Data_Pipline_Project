```python
create_table_query = '''
CREATE TABLE IF NOT EXISTS Raw_credits (
    person_id INT,
    id VARCHAR(50),
    name VARCHAR(500),
    character VARCHAR(500),
    role VARCHAR(100)
);
'''

cur.execute(create_table_query)
conn.commit()
```

```python
insert_row_3 = """
INSERT INTO Raw_credits (
    person_id, id, name, character, role
) VALUES (%s, %s, %s, %s, %s)
ON CONFLICT (id) DO NOTHING
"""

for _, row in Raw_credits.iterrows():
    # Ensure 'name', 'character', and 'role' are strings before checking length
    name = str(row['name'])[:255] if isinstance(row['name'], str) and len(row['name']) > 255 else row['name']
    character = str(row['character'])[:255] if isinstance(row['character'], str) and len(row['character']) > 255 else row['character']
    role = str(row['role'])[:255] if isinstance(row['role'], str) and len(row['role']) > 255 else row['role']

    # Prepare the row with the truncated values
    cur.execute(insert_row_3, (row['person_id'], row['id'], name, character, role))

conn.commit()
```
